```python
import pandas as pd
import numpy as np
```

```python
# load reported gene
reported = pd.read_csv('PD_gene_reported_anoted.csv')
```

```python
reported = reported.dropna()
```

```python
reported = reported.set_index('SYMBOL');
reported.head()
```

```python
reported.ENTREZID = reported.ENTREZID.astype(int)
```

```python
reportedGene = list(reported.ENTREZID)
print(len(reportedGene))
```

```python
#load DE
deData = pd.read_csv('OutputPD_limma.dat')
```

```python
deGenes = list(deData[deData['P.Value']<0.05]['ENTREZ_GENE_ID'])
print(len(deGenes))
```

```python
commonGene = list(set(deGenes).intersection(set(reportedGene)))
```

```python
reported.reset_index().set_index('ENTREZID').loc[commonGene]
```

```python
print(reported.reset_index().set_index('ENTREZID').loc[commonGene].SYMBOL)
```

```python
UnionGene = list(set(deGenes).union(set(reportedGene)))
```

```python
IntersectGene = list(set(deGenes).intersection(set(reportedGene)))
len(IntersectGene)
```
